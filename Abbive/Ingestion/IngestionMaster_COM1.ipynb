```python
import threading
import json
from multiprocessing.pool import ThreadPool,Pool
from datetime import datetime

FileStatus=[]
def run_notebook(inputParm):
    IOTconfigId, IOTsourceTypeId,USBconfigId,USBsourceTypeId,destinationContainerPath,destinationFolderPath,parent_run_id,job_id=inputParm
    PathDict={}
    PathDict['IOTconfigId']=IOTconfigId
    PathDict['IOTsourceTypeId']=IOTsourceTypeId
    PathDict['USBconfigId']=USBconfigId
    PathDict['USBsourceTypeId']=USBsourceTypeId
    PathDict['destinationContainerPath']=destinationContainerPath
    PathDict['destinationFolderPath']=destinationFolderPath
    PathDict['parent_run_id']=parent_run_id
    PathDict['job_id']=job_id
    # print(PathDict)
    return dbutils.notebook.run("./Ingestion_COM1", 900000, PathDict)
```

```python
pathlist=[]
parallelNotebook=10
for i in range(parallelNotebook):
    pathlist = pathlist+ [(IOTconfigId, IOTsourceTypeId,USBconfigId,USBsourceTypeId,destinationContainerPath,destinationFolderPath,parent_run_id,job_id)]
   
currentTime=datetime.utcnow()
print("TotalProcess:"+str(len(pathlist)))
pool = ThreadPool(parallelNotebook)
FileStatus.append(pool.map(run_notebook, pathlist))
pool.close()
pool.join()
messageProcessed=0
for i in FileStatus[0]:
    messageProcessed=messageProcessed+int(eval(i)[1])
processTime=datetime.utcnow()-currentTime
print("Time taken to process "+str(messageProcessed)+" is "+str(processTime.seconds/60)+" minute.")
print(str(messageProcessed/(processTime.seconds/60))+ " Messages per minute")
print("Completed for "+str("Logs_COM1_Raw"))
print("---------------------------------------------------------------")
```
