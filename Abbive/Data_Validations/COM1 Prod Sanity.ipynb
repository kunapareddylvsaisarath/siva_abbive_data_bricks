```sql
--FACT UE TABLE CT
-- From HdrFormatVersionCd to HdrDataContractNbr it is direct mapping to file (values will be directly in file)
-- 	ErrorDt to HdrAppHeadNbr it is direct mapping to file (values will be directly in file)
 select * from silverzone.FACTUELogs where FileNameDeviceTypeCd in ('CT')
 and HdrdateGeneratedDt='2020-09-25'  
 and InternalSerialNbr='CT4F1933601004'
 and ExternalSerialNbr='2020002552'
 AND HdrTimeGeneratedTmstmp='2020-02-11T08:57:53.313+0000' 
and HdrDataContractNbr like '21544%'
 and FileNameUUID = '05d690e8-16c0-47a4-959f-dd9333c0605a'
order by HdrTimeGeneratedTmstmp asc
```

```sql
--FACT UE TABLE COM3
-- From HdrFormatVersionCd to HdrDataContractNbr it is direct mapping to file (values will be directly in file)
-- 	From ZCodeMajorCd to RecoveryMessageTxt it is direct mapping to file (values will be directly in file)
-- 	HdrAppHeadNbr it is direct mapping to file (values will be directly in file)
select * from silverzone.FACTUELogs where FileNameDeviceTypeCd in ('COM3')
 and HdrdateGeneratedDt='2020-09-25'  
 and InternalSerialNbr='CSM41933601004'
 and ExternalSerialNbr='2020002552'
 AND HdrTimeGeneratedTmstmp='2020-02-11T08:57:53.313+0000' 
and HdrDataContractNbr like '21544%'
 and FileNameUUID = '05d690e8-16c0-47a4-959f-dd9333c0605a'
order by HdrTimeGeneratedTmstmp asc
```

```sql
-- To Check Duplicates on FACTUE 
select InternalSerialNbr,ExternalSerialNbr,HdrDataContractNbr,HdrTimeGeneratedTmstmp,count(1) as dup_count
FROM silverzone.FACTUELogs 
GROUP BY InternalSerialNbr,ExternalSerialNbr,HdrDataContractNbr,HdrTimeGeneratedTmstmp
HAVING COUNT(1) > 1;
```
